```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mplot
%matplotlib inline
import statsmodels.api as sm
import seaborn as sns


cars = sm.datasets.get_rdataset("cars", "datasets", cache=True)
# print(cars.__doc__)
print(cars.data.head())
data=pd.DataFrame(cars.data)
x1=data['speed']
y=data['dist']

x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
```

```python
#(a) Scatter plot: To Visualize the linear relationship between
#the predictor and response
mplot.ylabel('dist',fontsize=15)
mplot.xlabel('dist',fontsize=15)
mplot.scatter(x1,y)
```

```python
#Box plot
sns.boxplot(x='speed',y='dist',data=data,hue='speed', linewidth=2.5)
```

```python
#box plot
fig=mplot.figure(figsize=(10,7))
ax=fig.add_subplot(111)
bp=ax.boxplot(data)
```

```python
#Density plot
sns.distplot(data['speed'])
```

```python
# c) Print the summary statistics for linear the Mode
result.summary()
```

```python
# (b) Calculate correlation between speed and distance

mplot.ylabel('speed',fontsize=15)
mplot.xlabel('dist',fontsize=15)
mplot.scatter(x1,y)
y_hat=3.9324*x1+(-17.5791)
fig=mplot.plot(x1,y_hat,lw=2,c="red",label="regression line")
```
