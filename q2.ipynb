```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mplt
%matplotlib inline
import statsmodels.api as sm
import seaborn as sns

'''
The success of a shopping center can be represented as a function
of the distance (in km) from the center of the population and the
number of clients (in hundreds of people) who will visit the centre
each day. A set of collected data is given in the table below:
'''

No_of_Customers=[5,8,11,12,13,14,15,16,17]
distance=[12,8,7,8,4,3,2,1,0.5]
data_frame= pd.DataFrame.from_dict(dict([('No_of_Customers',No_of_Customers),('distance',distance)]))
data_frame
```

```python
y1=data_frame['No_of_Customers']
y1
mplt.ylabel('No_of_Customers',fontsize=15)
x1=data_frame['distance']
mplt.xlabel('distance',fontsize=15)
mplt.scatter(x1,y1)
```

```python
sns.boxplot(x='No_of_Customers',y='distance',data=data_frame,hue='No_of_Customers')
```

```python
sns.distplot(data_frame['distance'])
```

```python
x=data_frame['No_of_Customers']
y1=data_frame['distance']
y=sm.add_constant(y1)
result=sm.OLS(x,y).fit()
result.summary()
```

```python
yhat=-0.9542*y1+(17.1572)
yhat
```

```python
no_of_customer_3_miles=-0.9542*3+(17.1572)
print('no_of_customer_3_miles: ',no_of_customer_3_miles)
```
